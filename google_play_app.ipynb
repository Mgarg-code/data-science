```python
# removing the same data columns
df=df.drop('Genres',axis=1)
```

```python
df['Category'].describe()
```

```python
# which category having most app

a=df['Category'].value_counts()
#a.plot(kind='bar')
sns.countplot(df['Category'],data=df,palette="Set1")
plt.xticks(rotation=90)
plt.xlabel("CATEGORY",fontsize=20)
plt.ylabel("COUNTS",fontsize=20)
plt.title("show the category which contain maximum number of app in google play ",fontsize=22)
plt.show()
```

```python
df.describe()
```

```python
g = sns.catplot(x="Category",y="Rating",data=df, kind="box", height =25,
palette = "Set1")
g.despine(left=True)
g.set_xticklabels(rotation=90,fontsize=15)
g.set(yticks=range(0,5))
g.set( xticks=range(0,34))
g = g.set_ylabels("Rating",fontsize=15)
plt.title('Boxplot of Rating VS Category',fontsize = 20)
```

```python
# which app having mximum installations
sns.countplot(df['Installs'],data=df,palette='Set1')
plt.xticks(rotation=90)
plt.ylabel("COUNTS",fontsize=15)
plt.xlabel("NO OF INSTALLATIONS",fontsize=15)
plt.show()
# this graph show that mostally app have 1 milion + installations
```

```python
labels =df['Type'].value_counts(sort = True).index
sizes = df['Type'].value_counts(sort = True)

from pylab import rcParams
colors = ["palegreen","orangered"]
explode = (0.1,0)  # explode 1st slice
 
rcParams['figure.figsize'] = 8,8
# Plot
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=270,)

plt.title('Percent of Free App in store',size = 20)
plt.show()
```

```python
g = sns.catplot(x="Content Rating",y="Rating",data=df, kind="box", height = 10 ,palette = "Paired")
g.despine(left=True)
g.set_xticklabels(rotation=90)
g = g.set_ylabels("Rating")
plt.title('Box plot Rating VS Content Rating',size = 20)
```
